**Preparing data**

```python
import numpy as np
import pandas as pd

size = 300
X = np.random.rand(size)*5-2.5
w4, w3, w2, w1, w0 = 1, 2, 1, -4, 2
y = w4*(X**4) + w3*(X**3) + w2*(X**2) + w1*X + w0 + np.random.randn(size)*8-4
df = pd.DataFrame({'x': X, 'y': y})
df.to_csv('dane_do_regresji.csv',index=None)
df.plot.scatter(x='x',y='y')
```

**Test and train set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
df_train = pd.DataFrame({'x_train': X_train, 'y_train': y_train})
df_train.plot.scatter(x='x_train',y='y_train')
```

```python
df_test = pd.DataFrame({'x_test': X_test, 'y_test': y_test})
df_test.plot.scatter(x='x_test',y='y_test')
```

```python
X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

**Linear regression**

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
print(lin_reg.intercept_, lin_reg.coef_)
```

**KNN (k = 3)**

```python
import sklearn.neighbors
knn_reg_3 = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
knn_reg_3.fit(X_train, y_train)
print(knn_reg_3.predict([[0]]))
```

**KNN (k = 5)**

```python
import sklearn.neighbors
knn_reg_5 = sklearn.neighbors.KNeighborsRegressor(n_neighbors=5)
knn_reg_5.fit(X_train, y_train)
print(knn_reg_5.predict([[0]]))
```

**Polynomial Regression (deg 2)**

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features_2=PolynomialFeatures(degree=2,include_bias=False)
X_poly = poly_features_2.fit_transform(X_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y_train)
print(lin_reg_2.intercept_, lin_reg_2.coef_)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, alpha=0.5, label='Dataset')

p = np.poly1d(lin_reg_2.coef_[0])
x_2 = np.linspace(X.min(), X.max(), 400)
y_2 = p(x_2)

plt.plot(x_2, y_2)
```

**Polynomial Regression (deg 3)**

```python
poly_features_3=PolynomialFeatures(degree=3,include_bias=False)
X_poly_3 = poly_features_3.fit_transform(X_train)
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly_3, y_train)
print(lin_reg_3.intercept_, lin_reg_3.coef_)
```

```python
plt.scatter(X, y, alpha=0.5, label='Dataset')

p = np.poly1d(lin_reg_3.coef_[0])
x_3 = np.linspace(X.min(), X.max(), 400)
y_3 = p(x_3)

plt.plot(x_3, y_3)
```

**Polynomial Regression (deg 4)**

```python
poly_features_4=PolynomialFeatures(degree=4,include_bias=False)
X_poly_4 = poly_features_4.fit_transform(X_train)
lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly_4, y_train)
print(lin_reg_4.intercept_, lin_reg_4.coef_)
```

```python
plt.scatter(X, y, alpha=0.5, label='Dataset')

p = np.poly1d(lin_reg_4.coef_[0])
x_4 = np.linspace(X.min(), X.max(), 400)
y_4 = p(x_4)

plt.plot(x_4, y_4)
```

**Polynomial Regression (deg 5)**

```python
poly_features_5=PolynomialFeatures(degree=5,include_bias=False)
X_poly_5 = poly_features_5.fit_transform(X_train)
lin_reg_5 = LinearRegression()
lin_reg_5.fit(X_poly_5, y_train)
print(lin_reg_5.intercept_, lin_reg_5.coef_)
```

```python
plt.scatter(X, y, alpha=0.5, label='Dataset')

p = np.poly1d(lin_reg_5.coef_[0])
x_5 = np.linspace(X.min(), X.max(), 400)
y_5 = p(x_5)

plt.plot(x_5, y_5)
```

**Mean squared errors**

```python
from sklearn.metrics import mean_squared_error
mse_lin = mean_squared_error(y_train, lin_reg.predict(X_train))
mse_knn_3 = mean_squared_error(y_train, knn_reg_3.predict(X_train))
mse_knn_5 = mean_squared_error(y_train, knn_reg_5.predict(X_train))
mse_poly_2 = mean_squared_error(y_train, lin_reg_2.predict(poly_features_2.transform(X_train)))
mse_poly_3 = mean_squared_error(y_train, lin_reg_3.predict(poly_features_3.transform(X_train)))
mse_poly_4 = mean_squared_error(y_train, lin_reg_4.predict(poly_features_4.transform(X_train)))
mse_poly_5 = mean_squared_error(y_train, lin_reg_5.predict(poly_features_5.transform(X_train)))
```

```python
test_mse_lin = mean_squared_error(y_test, lin_reg.predict(X_test))
test_mse_knn_3 = mean_squared_error(y_test, knn_reg_3.predict(X_test))
test_mse_knn_5 = mean_squared_error(y_test, knn_reg_5.predict(X_test))
test_mse_poly_2 = mean_squared_error(y_test, lin_reg_2.predict(poly_features_2.transform(X_test)))
test_mse_poly_3 = mean_squared_error(y_test, lin_reg_3.predict(poly_features_3.transform(X_test)))
test_mse_poly_4 = mean_squared_error(y_test, lin_reg_4.predict(poly_features_4.transform(X_test)))
test_mse_poly_5 = mean_squared_error(y_test, lin_reg_5.predict(poly_features_5.transform(X_test)))
```

```python
mse = pd.DataFrame({
    'train_mse' : [mse_lin, mse_knn_3, mse_knn_5, mse_poly_2, mse_poly_3, 
                   mse_poly_4, mse_poly_5],
    'test_mse' : [test_mse_lin, test_mse_knn_3, test_mse_knn_5, test_mse_poly_2, 
                  test_mse_poly_3, test_mse_poly_4, test_mse_poly_5] },
    index=['lin_reg', 'knn_3_reg', 'knn_5_re',
            'poly_2_reg', 'poly_3_reg', 'poly_4_reg', 'poly_5_reg']
)

mse
```

```python
mse.to_pickle('mse.pkl')
```
