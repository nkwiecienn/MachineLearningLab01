Test and train set

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
```

```python
X, y = mnist.data, mnist.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Binary classifier

```python
y_train_0 = (y_train == '0')
y_test_0 = (y_test == '0')
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(n_jobs=-1)
sgd_clf.fit(X_train, y_train_0)
```

```python
y_train_pred = sgd_clf.predict(X_train)
y_test_pred = sgd_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

train_accuracy = accuracy_score(y_train_0, y_train_pred)
test_accuracy = accuracy_score(y_test_0, y_test_pred)
```

```python
train_accuracy
```

```python
test_accuracy
```

```python
import pickle

with open('sgd_acc.pkl', 'wb') as f:
    pickle.dump([train_accuracy, test_accuracy], f)
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(sgd_clf, X_train, y_train_0, cv=3, scoring='accuracy', n_jobs=-1)
cv_scores
```

```python
with open("sgd_cva.pkl", 'wb') as f:
    pickle.dump(cv_scores, f)
```

Whole set classifier

```python
sgd_clf.fit(X_train, y_train)
y_test_pred = sgd_clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

conf_mx = confusion_matrix(y_test, y_test_pred)
conf_mx
```

```python
with open("sgd_cmx.pkl", 'wb') as f:
    pickle.dump(conf_mx, f)
```
